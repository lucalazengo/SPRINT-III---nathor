```python
df2 = df2.rename(columns={
    'Codigo Produto': 'CODIGO_PRODUTO',
    'Descricao Produto': 'DESCRICAO_PRODUTO'
})
# Definir o intervalo de datas de abril
inicio = pd.Timestamp('2024-04-01')
fim = pd.Timestamp('2024-04-30')

# Gerar datas aleatórias
df2['Data_Entrega'] = np.random.choice(
    pd.date_range(start=inicio, end=fim),
    size=len(df2)
)
df2['Data_Entrega'] = df2['Data_Entrega'].dt.strftime('%d/%m/%Y')

# Salvar em CSV
df2.to_csv(
    '../data/processed/pedidos.csv',
    index=False,
    encoding='utf-8'
)
# Salvar em JSON
df2.to_json(
    '../data/processed/pedidos.json',
    orient='records',
    force_ascii=False
)
```
